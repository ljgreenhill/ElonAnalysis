```python
import yfinance as yf
import pandas as pd
import datetime
```

```python
# Retrieve TSLA Stock
tsla = yf.Ticker("TSLA")

# Show stock history
tsla_hist = tsla.history(period="max")
tsla_hist.head()
```

```python
# Create column for change in price
# Change in price = Close price - Open price
tsla_hist["Change In Price"] = tsla_hist["Close"] - tsla_hist["Open"]
tsla_hist.head()
```

```python
# Show Index of dataframe
tsla_hist.index
```

```python
# Save only relevant dates
tsla_data = tsla_hist.loc['2020-01-25':, ['Change In Price']]
tsla_data.head()
```

```python
# Read in twitter data csv
twitter_data_numeric = pd.read_csv('data/tweets.csv', usecols=['Favorites', 'Retweets', 'Date Created'])
twitter_data_text = pd.read_csv('data/tweets.csv', usecols=['Date Created', 'Text'])
twitter_data_numeric.head()
twitter_data_text.head()
```

```python
#remove hours, minutes, and seconds from date
twitter_data_numeric["Date Created"] = twitter_data_numeric["Date Created"].str.slice(0, 10)
twitter_data_text["Date Created"] = twitter_data_text["Date Created"].str.slice(0, 10)
twitter_data_numeric.head()
```

```python
#find average of retweet and favorites by day per tweet
twitter_data_numeric = twitter_data_numeric.groupby(["Date Created"]).mean()
twitter_data_numeric.head()
```

```python
#combine text of all tweets in one day
twitter_data_text = twitter_data_text.groupby(["Date Created"])["Text"].apply(','.join).reset_index()
twitter_data_text.head()
```

```python
#create one tweet dataframe
twitter_data_numeric.reset_index(drop=True, inplace=True)
twitter_data_text.reset_index(drop=True, inplace=True)
dfs = [twitter_data_numeric, twitter_data_text]
twitter_data = pd.concat(dfs, axis=1)
twitter_data = twitter_data.set_index('Date Created')
twitter_data.head()
```

```python
#Remove dates when Elon did not tweet or there is no stock data
tsla_data.index = tsla_data.index.astype(str)
tsla_data.index = tsla_data.index.str.slice(0, 11)
for date in tsla_data.index:
    if date not in twitter_data.index:
        tsla_data = tsla_data.drop(date)
for date in twitter_data.index:
    if date not in tsla_data.index:
        twitter_data = twitter_data.drop(date)
print(tsla_data.shape)
print(twitter_data.shape)
```

```python
#Combine both datasets
output = pd.concat([tsla_data, twitter_data], axis = 1)
output.head()
```

```python
# Save data as csv file
output.to_csv('data/elon.csv')
```
