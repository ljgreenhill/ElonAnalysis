```python
import yfinance as yf
import pandas as pd
import datetime
```

```python
# Retrieve TSLA Stock
tsla = yf.Ticker("TSLA")

# Show stock history
tsla_hist = tsla.history(period="max")
tsla_hist.head()
```

```python
# Create column for change in price
# Change in price = Close price - Open price
tsla_hist["Change In Price"] = tsla_hist["Close"] - tsla_hist["Open"]
tsla_hist.head()
```

```python
# Show Index of dataframe
tsla_hist.index
```

```python
# Save only relevant dates
tsla_data = tsla_hist.loc['2020-01-25':, 'Change In Price']
tsla_data.head()
```

```python
# Read in twitter data csv
twitter_data = pd.read_csv('data/tweets.csv', usecols=['Favorites', 'Retweets', 'Date Created'])
twitter_data.head()
```

```python
#remove hours, minutes, and seconds from date
print(type(twitter_data["Date Created"]))
twitter_data["Date Created"] = twitter_data["Date Created"].str.slice(0, 11)
twitter_data.head()
```

```python
#find average of retweet and favorites by day per tweet
twitter_data = twitter_data.groupby(["Date Created"]).mean()
twitter_data.head()
```

```python
#Combine both datasets
print(twitter_data.shape)
print(tsla_data.shape)
dataset = pd.concat([tsla_data, twitter_data], axis=1, sort=False)
dataset.head(3)
```

```python
# Save data as csv file
tsla_data.to_csv('data/tesla_data.csv')
```
